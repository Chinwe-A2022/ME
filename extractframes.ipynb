```python
# Import the following packages
import cv2 # Represent OpenCV library for capturing videos or images
import os # To save or open a file directory on the ocmputer

# Call OpenCV to capture image sequence from video files
cap = cv2.VideoCapture(r'G:\Box June\BW_JUNE20220611\1\N844A8_ch1_main_20220611073000_20220611080000.mp4')

# Obtaining the number of frames and fps to calculate the total duration
number_of_frames = cap.get(cv2.CAP_PROP_FRAME_COUNT)
print(number_of_frames)
frames_per_second = cap.get(cv2.CAP_PROP_FPS)
print(frames_per_second)
total_duration = (number_of_frames// frames_per_second)
print(total_duration)

# Define a function to capture 1fps
def getFrame(sec):
    cap.set(cv2.CAP_PROP_POS_MSEC, sec*10000)
    ret, frame = cap.read()  
    return frame;

a = 760
sec = 0

 # Set the while loop for the following functions to be executed while the total video duration is not elapsed
while sec <= total_duration:
     frame = getFrame(sec)
   
 
     # It will capture image in ever 1 second        
     frameRate = 1  
     sec = sec + frameRate
     sec = round(sec, 2)
     print (len(frame))

     path = r'F:\BOXWOOD_MEDIUM'
     cv2.imwrite(os.path.join(path, str(a) + '.png'), frame)
     a +=1
     print(a)
```
